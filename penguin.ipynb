```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
sns.scatterplot(data=penguins,x='bill_length_mm',y='bill_depth_mm',hue = 'sex')
plt.title("bill_length vs bill_depth")
```

```python
sns.pairplot(penguins,hue='species')
```

```python
sns.boxplot(data=penguins,x='sex',y='bill_length_mm')
```

```python
penguins_num = penguins.select_dtypes(include=['int64','float64'])
penguins_num.head()
```

```python
penguins_num.corr()
```

```python
corr_matrix = penguins_num.corr()
sns.heatmap(corr_matrix,annot = True,cmap='coolwarm')
```

```python
sns.violinplot(data=penguins,x='sex',y='bill_length_mm')
```

```python
sns.countplot(data=penguins)
```

```python
sns.stripplot(data=penguins,x='sex',y='bill_length_mm')
```

```python
sns.swarmplot(data=penguins,x='sex',y='bill_length_mm')
```

```python
sns.kdeplot(data=penguins, x='bill_length_mm', y='bill_depth_mm')
plt.title("KDE Plot of bill Length vs bill Depth")
plt.show()
```
